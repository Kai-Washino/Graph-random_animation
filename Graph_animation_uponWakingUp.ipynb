```python
%matplotlib notebook
 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import random

fig = plt.figure()
 
xlim = [0,100]
X, Y = [], []
Normal_time = 30
Change_time = 100
Change_time2 = 50
finish_time = 50
Change_rate =50
range1 = 60
range2 = 75

def plot(data):
    plt.cla()    # 前のグラフを削除
    if len(Y) < Normal_time: #Nomarl_timeまで到達していないとき
        Y.append(random.randrange(range1,range2))   # データを作成
        X.append(len(Y))
    elif (Normal_time <= len(Y)) and (len(Y) < Normal_time + Change_time):#Normal＿timeに到達したとき
        Y.append(random.randrange(range1+Change_rate,range2+Change_rate))   # データを作成
        X.append(len(Y))
    elif (Normal_time + Change_time <= len(Y)) and (len(Y) < Normal_time + Change_time + Change_time2):
        Y.append(random.randrange(range1-len(Y) + Normal_time + Change_time ,range2-len(Y) + Normal_time + Change_time ))   # データを作成
        X.append(len(Y))
    else :#Change＿timeに到達したとき
        Y.append(random.randrange(range1,range2))   # データを作成
        X.append(len(Y))
    if len(X) > 100:            # 描画範囲を更新
        xlim[0]+=1
        xlim[1]+=1
    plt.plot(X, Y)              # 次のグラフを作成
    plt.title("Saliva volume")
    plt.ylim(0,150)
    plt.xlim(xlim[0],xlim[1])
    
    # 10msごとにplot関数を呼び出してアニメーションを作成

ani = animation.FuncAnimation(fig, plot, interval=10, blit=True, frames = 430 )
ani.save('UW.gif', writer='pillow')
plt.show()
```
